```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
archEvent=pd.read_csv('/home/sherly/Escritorio/fiuba-trocafone-tp1-final-set/events.csv',low_memory=False)
archEvent.head(5)
```

```python
archEvent.size
```

```python
tipoEvent_Cant=archEvent['event'].value_counts()
tipoEvent_Cant
```

# Ejercicio 1: Analizar cuál es el tipo de evento predominante en el set de datos. 


```python
f = plt.figure()

grafhist=tipoEvent_Cant

plt.title('Cantidad de eventos en el set de datos ', color='black',fontsize=15)

ax=grafhist.plot(kind='bar',rot=80,figsize=(15,12),fontsize=14);

ax.set_ylabel('Cantidad de eventos',fontsize=16);
ax.set_xlabel('Tipo de eventos',fontsize=16);


```

# Ejercicio 2: Calcular cuales son los 5 productos (identificados por sku) con mayor cantidad de conversiones.

```python
#tipo de evento y sku(producto)
evento_producto=archEvent[['event','sku']]
evento_producto.head(5)
```

```python
conversion_producto=evento_producto.loc[evento_producto['event']=='conversion']
cant_converproducto=conversion_producto['sku'].value_counts()
top5Converproducto=cant_converproducto.head(5)
```

```python
top5Converproducto
```

```python
f = plt.figure()

grafhist = top5Converproducto

plt.title('Top 5 productos con mayor cantidad de conversion ', color='black',fontsize=16)

ax=grafhist.plot(kind='bar',rot=80,figsize=(15,12),color=['r','g','k','c','b'],fontsize=14);
ax.set_ylabel('cantidad de conversion',fontsize=16);
ax.set_xlabel('identificador del producto (sku)',fontsize=16);
```

# Ejercicio3: Averiguar cuáles son los 10 usuarios que registraron más actividad en la plataforma (es decir, realizaron más eventos).

```python
cantEventos_cliente=archEvent['person'].value_counts()
top10Clientes_eventos=cantEventos_cliente.head(10)
top10Clientes_eventos
```

```python
f = plt.figure()

grafhist = top10Clientes_eventos

plt.title('Top 10 usuarios que registran mayor actividad ', color='black',fontsize=16)

ax=grafhist.plot(kind='bar',rot=80,figsize=(15,12),fontsize=14);
ax.set_ylabel('cantidad de eventos',fontsize=16);
ax.set_xlabel('clave de usuarios',fontsize=16);
```
